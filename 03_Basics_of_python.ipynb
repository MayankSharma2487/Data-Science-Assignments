# **Prime Numbers**

```python
def prime_number_check(number):
  if number <= 1:
    print('Not prime ')
  elif number == 2:
    print('Prime')
  else:
    for i in range(2,number):
      if number % i == 0:
        print('Not Prime')
        return
    print('Prime')
```

```python
prime_number_check()
```

# **Product of Numbers**

```python
import numpy as np
def random_numbers(n):
  numbers = np.random.randint(1,n,2)
  return numbers
```

```python
numbers = random_numbers(10)
print('your numbers are:- ',numbers)
product = int(input('Enter The product :-'))
if product == np.multiply(numbers[0],numbers[1]):
  print('Correct')
else:
  print('Incorrect! product is :- ',np.multiply(numbers[0],numbers[1]))
```

```python
def match_product():
  numbers = random_numbers(10)
  print('your numbers are:- ',numbers)
  product = int(input('Enter The product :-'))
  if product == np.multiply(numbers[0],numbers[1]):
    print('Correct')
  else:
    print('Incorrect! product is :- ',np.multiply(numbers[0],numbers[1]))
```

```python
match_product()
```

# **Word Counter**

```python
text = "hello hello My name is Mayank, How are you doing, What are you doing, Why are you doing"
count = {}

cleaned_text = text.lower().replace(',', '')

for word in cleaned_text.split():
    if word in count:
        count[word] += 1
    else:
        count[word] = 1

print(count)
```

## **Check for Palindrome**

```python
text = 'Racecar'
text = text.lower()
if text[::-1] == text:
   print('True')
else:
  print('False')
```

```python
def palindrome_check():
  text = str(input('Enter Word :- '))
  text = text.lower()
  if text[::-1] == text:
    print('True')
  else:
    print('False')
```

```python
palindrome_check()
```
