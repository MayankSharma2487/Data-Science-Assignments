````python
import numpy as np
from scipy.stats import chi2_contingency, chi2

obs = np.array([
    [50, 70],   # Very Satisfied
    [80,100],   # Satisfied
    [60, 90],   # Neutral
    [30, 50],   # Unsatisfied
    [20, 50]    # Very Unsatisfied
])

chi2_stat, p_value, dof, expected = chi2_contingency(obs)

alpha = 0.05
critical = chi2.ppf(1 - alpha, dof)

print("Observed:\n", obs)

````

````python
print("\nExpected:\n", expected.round(3))
````

````python
print(f"\nChi-square statistic = {chi2_stat:.6f}")
````

````python
print(f"Degrees of freedom = {dof}")
````

````python
print(f"p-value = {p_value:.6f}")
````

````python
print(f"Critical value (alpha={alpha}) = {critical:.3f}")
````

````python
if chi2_stat > critical:
    print("\nDecision: Reject H0 -> There IS an association.")
else:
    print("\nDecision: Fail to reject H0 -> NO evidence of association.")
````
