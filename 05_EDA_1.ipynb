# **Outlier Handling**

```python
df.boxplot()
```

```python
def outliers(data):
  data_num = data.select_dtypes(include = 'number')

  for col in data_num:
      q1 = data[col].quantile(0.25)
      q3 = data[col].quantile(0.75)

      IQR = q3-q1
      lower_bound = q1 - 1.5*IQR
      upper_bound = q3 + 1.5*IQR

      data[col] = np.where(data[col] < lower_bound, lower_bound,
                      np.where(data[col] > upper_bound, upper_bound, data[col]))

  return data

```

```python
df_new = outliers(df)
```

### **Violin Plot**

```python
df_new.columns
```

```python

cols = ['LB', 'AC', 'FM', 'UC', 'DL', 'ASTV', 'MSTV', 'ALTV', 'MLTV', 'Width',
       'Tendency']

for col in cols:
  sns.violinplot(data=df_new[col])
  plt.title(f'Violin Plot of {col}')
  plt.xlabel(col)
  plt.show()
```

### **Insights from Violin Plots**


*  **LB :-** the median is around 130, 50 % of the data falls Between 120 and 140, data is more dense around the median
*   **AC and UC :-** both plots have a median close to 0. The violin is very sharp around the 0, it signifies that most of the values are around the 0
* **FM and DL :-** The median is around the 0, similar to AC and UC. there are some values that spread out to both +ve and -ve extreams.
* **ASTV :-** distribution is somewhat symmetrical but with a wider, shape at the top and bottom. The median is around 50. The width of the plot indicates a relatively even spread of values between 20 and 80.

* **MSTV :-** This plot is similar to ASTV in that it shows a wide, even distribution of values, but its scale is much smaller. The median appears to be around 1.
* **MLTV :-** This plot has a median around 7-8 and shows that the data is concentrated around this value. The long tails indicate a spread of values, with some outliers, but most are close to the center.
* **Width :-** The Width plot is very similar to ASTV in its overall shape and distribution. It shows a central tendency around the median of about 70 and a wide spread of values.

* **Tendency :-** The Tendency plot is the most unique. It clearly shows a trimodal distribution, with three distinct peaks in density. This suggests the data for Tendency is clustered around three separate values, likely -1, 0, and 1. This means the variable has a high probability of taking on one of these three distinct values.
