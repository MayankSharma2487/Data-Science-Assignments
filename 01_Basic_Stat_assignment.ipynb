## Bar Chart Analysis for Categorical Column

```python
df.info()
```

```python
cat_df=df.select_dtypes('object')
cat_df
```

```python
sns.set(style="whitegrid")

cat_cols = ['Day', 'City', 'BU', 'Brand']

for col in cat_cols:
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette='Set2')
    plt.title(f'Frequency of {col}', fontsize=14)
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```

### Analysis Based on the Barcharts 
+ Thursday has more sales data 90 compared to other days 60.
+ All records are from a single city (C).
+ Perfectly balanced distribution across all three BU.
+ Jeera leads in frequency, followed by Gear and Viva

```python
import numpy as np
def z_score(x):
    n = len(x)
    mu = np.mean(x)
    var = (np.sum((x-mu)**2)/n)
    sigma = np.sqrt(var)
    zscore = (x-mu)/sigma
    return zscore
    
```

### Bfore Standardization

```python
print("""Standard Deviation :
""",df_numerical.std())
print()
print("""Min Values
""",df_numerical.min())
print()
print("""Max Values
""",df_numerical.max())
print()
print("""Mean
""",df_numerical.mean())
```

### After Standardization

```python
standardized_df = df_numerical.apply(z_score)
print("""Standard Deviation :
""",standardized_df.std())
print()
print("""Min Values
""",standardized_df.min())
print()
print("""Max Values
""",standardized_df.max())
print()
print("""Mean
""",standardized_df.mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']

for col in cols:
    fig, axs = plt.subplots(1, 2, figsize=(14, 5))
    

    sns.kdeplot(df[col], ax=axs[0], color='blue', fill=True)
    axs[0].set_title(f'Original Distribution - {col}')
    axs[0].grid(True)


    sns.kdeplot(standardized_df[col], ax=axs[1], color='red', fill=True)
    axs[1].set_title(f'Standardized Distribution - {col}')
    axs[1].grid(True)

    plt.tight_layout()
    plt.show()
```
